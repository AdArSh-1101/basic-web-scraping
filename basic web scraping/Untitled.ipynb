```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(r.text, "lxml")
print(soup)
```

```python
boxes=soup.find_all("div",class_="col-md-4 col-xl-4 col-lg-4")
print(len(boxes))
```

```python
names=soup.find_all("a",class_="title")
print(names)
```

```python
n=[]
p=[]
d=[]
for i in names:
    n.append(i.text)
```

```python
prices=soup.find_all("h4",class_="float-end price card-title pull-right")
print(prices)
```

```python
for i in prices:
    p.append(i.text)
```

```python
desc=soup.find_all("p",class_="description card-text")
for i in desc:
    d.append(i.text)
```

```python
import numpy as np
import pandas as pd
```

```python
data=list(zip(n,p,d))
```

```python
df=pd.DataFrame(data ,columns=['Name','Price','Description'])
df
```

```python
df.to_csv('output.csv',index=False)
```
